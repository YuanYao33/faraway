```python
#input 1-day data
import pandas as pd
import numpy as np
url = "AMF_US-Ho2_FLUXNET_DD_2007-2020.csv" # 1-day intervals

data_1_day = pd.read_csv(url)

#Read the raw 1-day data
PAR_DD_df = data_1_day['PPFD_IN']
PAR_DD_df[PAR_DD_df < 0] = np.nan # Turn -9999 of PAR_DD to "Nan"
PAR_DD = PAR_DD_df.tolist()

TA_DD = data_1_day['TA_F'].tolist()

GPP_DD = data_1_day['GPP_DT_VUT_REF'].tolist()

Precipitation_DD = data_1_day['P_F'].tolist()

#Convert the raw 1-day to 8-day data.
#Calculate the eight-day average (NaN does not participate in the average).
#Note: DataUnit conversion, e.g. convert from per second to per day
PAR_8D = []
TA_8D = []
GPP_8D = []
Precipitation_8D = []
day_num = 1

#From Year 2007 to 2019
for i in range(13):    
    for j in range(46):
        day_start = day_num + 8 * j
        day_end = day_start + 7                                            
        PAR_8D.append(np.nanmean(PAR_DD[day_start : day_end]) * 0.000001 * 60 * 60 * 24) # Units converted from umol m-2 s-1 to mol m-2 d-1          
        GPP_8D.append(np.nanmean(GPP_DD[day_start : day_end])) 
        TA_8D.append(np.nanmean(TA_DD[day_start : day_end]))
        Precipitation_8D.append(np.nansum(Precipitation_DD[day_start : day_end]))
        
    day_num = day_num + 365 # Treat all years as simply 365 days (ignoring 366 days)

#Year 2020
#The last eight-day time period of the final year is less than eight days
for j in range(45):
    day_start = day_num + 8 * j
    day_end = day_start + 7                                            
    PAR_8D.append(np.nanmean(PAR_DD[day_start : day_end]) * 0.000001 * 60 * 60 * 24) # Units converted from umol m-2 s-1 to mol m-2 d-1          
    GPP_8D.append(np.nanmean(GPP_DD[day_start : day_end]) )
    TA_8D.append(np.nanmean(TA_DD[day_start : day_end]))
    Precipitation_8D.append(np.nansum(Precipitation_DD[day_start : day_end]))

day_start = day_num + 8 * 45
day_end = len(PAR_DD)
PAR_8D.append(np.nanmean(PAR_DD[day_start : day_end]) * 0.000001 * 60 * 60 * 24) # Units converted from umol m-2 s-1 to mol m-2 d-1          
GPP_8D.append(np.nanmean(GPP_DD[day_start : day_end])) 
TA_8D.append(np.nanmean(TA_DD[day_start : day_end]))
Precipitation_8D.append(np.nansum(Precipitation_DD[day_start : day_end]))

data_8_day = pd.read_csv('8-day_DATA_US-Ho2_2007_2020_Timeline.csv')
data_8_day['PAR_8D'] = PAR_8D
data_8_day['GPP_8D'] = GPP_8D
data_8_day['TA_8D'] = TA_8D
data_8_day['Precipitation_8D'] = Precipitation_8D

print(data_8_day)
data_8_day.to_csv("8-day_DATA_US-Ho2_2007_2020_From_1-day.csv",index=False,sep=',')
```
