```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd 


### reading data ###
file = '8-day_DATA_US-Ho2_2007_2020_ForPY_Calculate.xlsx'
data = pd.read_excel(file)


#ETec = data['ETec']
#TRvtm = data['TRvtm']
GPPvpm = data['GPPvpm_Calculate']
GPPec = data['GPP_8D']
GPPmod17 = data['Mod17_GPP']
#ETmodel = data['ET_MODIS']


#scatter1

#ETec = ETec[~np.isnan(TRvtm)]
#TRvtm = TRvtm[~np.isnan(TRvtm)]

#ETec = list(ETec)
#TRvtm = list(TRvtm)
#ETec_new = []
#TRvtm_new = []
#ETec_new1 = []
#TRvtm_new1 = []


#scatter2

GPPec = GPPec[~np.isnan(GPPmod17)]
GPPmod17 = GPPmod17[~np.isnan(GPPmod17)]
GPPmod17 = GPPmod17[~np.isnan(GPPec)]
GPPec = GPPec[~np.isnan(GPPec)]

GPPec = list(GPPec)
GPPmod17 = list(GPPmod17)

GPPec_new = []
GPPmod17_new = []
GPPec_new1 = []
GPPmod17_new1 = []

#scatter3


#GPPec = GPPec[~np.isnan(GPPvpm)]
#GPPvpm = GPPvpm[~np.isnan(GPPvpm)]
#GPPvpm = GPPvpm[~np.isnan(GPPec)]
#GPPec = GPPec[~np.isnan(GPPec)]

#GPPec = list(GPPec)
#GPPvpm = list(GPPvpm)

#GPPec_new = []
#GPPvpm_new = []
#GPPec_new1 = []
#GPPvpm_new1 = []





#scatter4
#ETec = ETec[~np.isnan(ETmodel)]
#ETmodel = ETmodel[~np.isnan(ETmodel)]
#ETmodel = ETmodel[~np.isnan(ETec)]
#ETec = ETec[~np.isnan(ETec)]

#ETec = list(ETec)
#ETmodel = list(ETmodel)
#ETec_new = []
#ETmodel_new = []
#ETec_new1 = []
#ETmodel_new1 = []




for index,value in enumerate(GPPec):
    if(value > 0):
        GPPec_new.append(GPPec[index])
        GPPmod17_new.append(GPPmod17[index])

for index,value in enumerate(GPPmod17_new):
    if(value > 0):
        GPPec_new1.append(GPPec_new[index])
        GPPmod17_new1.append(GPPmod17_new[index]) 


p=np.polyfit(GPPec_new1,GPPmod17_new1,1)

pp1 = "{:.4f}".format(p[0])
pp2 = "{:.4f}".format(p[1])
print(pp1)
print(pp2)

h=np.corrcoef(GPPec_new1,GPPmod17_new1)
hh = h[0,1]*h[0,1]
print(hh)

### making graphs ###
#scatter

fig, ax = plt.subplots(num=1)

x=range(20);
y=range(20);

X =  np.arange(0,15,0.1)
Y = p[0]*X+p[1]

plt.scatter(GPPec_new1,GPPmod17_new1,color='white', marker='o', edgecolors='b',s=20)
plt.plot(x,y,'r--',linewidth=2.5)
plt.plot(X,Y,'k-',linewidth=2.5)




#ax.set_xlabel('ETec(mm/day)8-day PARgt0 average', fontsize=12)
ax.set_xlabel('GPPec(g C/m2/day)', fontsize=12)
ax.set_ylabel('GPPmod17(mm/day)', fontsize=12)
#ax.set_ylabel('TRvpm(mm/day) 8-day PARgt0 average', fontsize=12)
#ax.set_ylabel('GPPvpm(g C/m2/day)8-day PARgt0 average', fontsize=12)
#ax.set_ylabel('GPPvpm(g C/m2/day)8-day PARgt0 average', fontsize=12)

plt.rcParams['xtick.direction']='in'
plt.rcParams['ytick.direction']='in'

plt.xlim((0,16))
plt.ylim((0,16))

plt.xticks([0,2,4,6,8,10,12,14,16],fontsize=12)
plt.yticks([0,2,4,6,8,10,12,14,16],fontsize=12)

#plt.xticks([0,0.5,1,1.5,2,2.5,3,3.5,4,4.5,5],fontsize=12)
#plt.yticks([0,0.5,1,1.5,2,2.5,3,3.5,4,4.5,5],fontsize=12)
#plt.xticks([0,2,4,6,8,10,12,14,16],fontsize=12)
#plt.yticks([0,2,4,6,8,10,12,14,16],fontsize=12)

plt.text(1,14,'R2=%0.6s' %hh,fontsize=13)
plt.text(4.7,14,'y='+str(pp1)+'x'+'+'+str(pp2),fontsize=13)

#plt.savefig('result_12_PARgt0_Scatter_GPPec_GPPvpm_Howland_2007_2020.png')
#plt.savefig('PARgt0_Scatter_ETec_ECmod16_Howland_2007_2020.png')
#plt.savefig('PARgt0_Scatter_ETec_TRvtm_Howland_2007_2020.png')
plt.savefig('result_11_PARgt0_Scatter_GPPec_GPPmod17_Howland_2007_2020.png')
```
