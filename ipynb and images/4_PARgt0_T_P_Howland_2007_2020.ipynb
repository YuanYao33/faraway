```python
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1 import host_subplot
from mpl_toolkits import axisartist
import numpy as np
import pandas as pd 

### reading data ###
file = '8-day_DATA_US-Ho2_2007_2020_ForPY_Calculate.xlsx'
data = pd.read_excel(file)


T=data['TA_8D']
P=data['Precipitation_8D']
PAR = data['PAR_8D']
X = range(len(T))


### making graphs ###
#PARgt0_T_P_Howland_2007_2020

cc = '#1E90FF'
fig, ax1 = plt.subplots(num=2,figsize = (20,10))

plt.rcParams['xtick.direction']='in'
plt.rcParams['ytick.direction']='in'

ax2 = ax1.twinx()
ax3 = ax1.twinx()

l1 =ax1.plot(X,T,color='r', marker='*',linestyle='-',markersize=4,linewidth=0.5, label='Air temperature()')
l3 =ax2.plot(X,P,color='blue', marker='d',markerfacecolor='white',linestyle='-',markersize=4,linewidth=0.5, label='Precipitation(mm/day)')
l4 =ax3.plot(X,PAR,color='k',markersize=4, label='PAR(mol*m-2*d-1)')

left2 = ax2.spines['left'].set_color('r')
right = ax2.spines['right'].set_color('blue')
right2 = ax3.spines['right'].set_position(('outward', 60))
right2 = ax3.spines['right'].set_color('k')


ax1.set_xlabel('Date', fontsize=14)
ax1.set_ylabel('Air temperature() ', fontsize=14)
ax2.set_ylabel('Precipitation(mm/day) ', fontsize=14)
ax3.set_ylabel('PAR(mol*m-2*d-1)', fontsize=14)

ax1.yaxis.label.set_color('r')
ax2.yaxis.label.set_color('blue')
ax3.yaxis.label.set_color('k')

ax1.set_xlim((0,len(X)))
ax1.set_ylim((-20,35))
ax2.set_ylim((0,14*8))

ax1.set_xticks([1,47,93,139,185,231,277,323,369,415,461,508,554,599,645],\
           [2007,2008,2009,2010,2011,2012,2013,2014,2015,2016,2017,2018,2019,2020,2021],fontsize=14)

ax1.set_yticks([-20,-15,-10,-5,0,5,10,15,20,25,30,35],fontsize=16,colors='r')
ax2.set_yticks([0,16,32,48,64,80,96,112],fontsize=16,colors='blue')
ax3.set_yticks([0,10,20,30,40,50,60],fontsize=16,colors='k')

for tl in ax1.get_yticklabels():
    tl.set_color('r')
for tl in ax2.get_yticklabels():
    tl.set_color('blue')
for tl in ax3.get_yticklabels():
    tl.set_color('k')
    
ax1.tick_params(axis='y',colors='r')
ax2.tick_params(axis='y',colors='blue')  
ax3.tick_params(axis='y',colors='k') 
    
lns = l1+l3+l4
labs = [l.get_label() for l in lns]

ax2.legend(lns, labs,loc ='upper left',prop={"size":12})

plt.savefig('result_2_PARgt0_T_P_PARHowland_2007_2020.png')




















```
