```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd 


### reading data ###
file = '8-day_DATA_US-Ho2_2007_2020_ForPY_Calculate.xlsx'
data = pd.read_excel(file)

GPPvpm = data['GPPvpm_Calculate']
GPPec = data['GPP_8D']
GPPmod17 = data['Mod17_GPP']

X = range(len(GPPvpm))


### making graphs ###
#GPPmod17_GPPec_GPPvpm_Howland_2007_2020
fig, ax = plt.subplots(num=1,figsize = (12,6))

ax.plot(X,GPPec,color='r', marker='*',linestyle='-',markersize=4,linewidth=0.5, label='GPPec')
ax.plot(X,GPPvpm,color='blue', linestyle='-',markersize=3,linewidth=0.5, label='GPPvpm')
ax.plot(X,GPPmod17,color='yellow', linestyle='-',markersize=4,linewidth=0.5, label='GPPmod17')

plt.xlim((0,len(GPPvpm)))
plt.ylim((0,18))

ax.set_xlabel('Year', fontsize=14)
ax.set_ylabel('GPPec, GPPvpm and GPPmod17(g C/m2/day)', fontsize=12)

plt.rcParams['xtick.direction']='in'
plt.rcParams['ytick.direction']='in'

plt.xticks([1,47,93,139,185,231,277,323,369,415,461,508,554,599,645],\
           [2007,2008,2009,2010,2011,2012,2013,2014,2015,2016,2017,2018,2019,2020,2021],fontsize=12)
plt.yticks([0,2.5,5,7.5,10,12.5,15,18],fontsize=12)

leg= ax.legend(prop={"size":11},loc = 'best')
plt.savefig('result_10_GPPmod17_GPPec_GPPvpm_Howland_2007_2020.png')





























































```
