```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd 

### reading data ###
file = '8-day_DATA_US-Ho2_2007_2020_ForPY_Calculate.xlsx'
data = pd.read_excel(file)

GPPec = data['GPP_8D']
#NEEec = data['NEE_VUT_REF_Unit_gC_HH']
#ETec = data['ETec']
X = range(len(GPPec))

### making graphs ###
#PARgt0_GPPec_NEEec_ETec_Howland_2007_2020
cc = 'blue'
fig, ax1 = plt.subplots(num=2,figsize = (12,6))

plt.rcParams['xtick.direction']='in'
plt.rcParams['ytick.direction']='in'
#ax2 = ax1.twinx()

l1 =ax1.plot(X,GPPec,color=cc, marker='*',markerfacecolor='white',linestyle='-',markersize=8,linewidth=0.5, label='GPPec')

#l2 =ax1.plot(X,-NEEec,color='g', marker='o',markerfacecolor='white',linestyle='-',markersize=4,linewidth=0.5, label='NEEec')
#l3 =ax2.plot(X,ETec,color='fuchsia', marker='d',markerfacecolor='white',linestyle='-',markersize=4,linewidth=0.5, label='ETec')

#left2 = ax2.spines['left'].set_color('C0')
#right = ax2.spines['right'].set_color('fuchsia')

ax1.set_xlim((0,len(X)))
ax1.set_ylim((0,16))
#ax2.set_ylim((0,4.5))

ax1.set_xlabel('Date', fontsize=14)
ax1.set_ylabel('GPPec(gC*m-2*d-1)', fontsize=14)
#ax2.set_ylabel('ETec(mm/day) 8-day PARgt0 average', fontsize=14)
ax1.yaxis.label.set_color(cc)
#ax2.yaxis.label.set_color('fuchsia')

ax1.set_xticks([1,47,93,139,185,231,277,323,369,415,461,508,554,599,645],\
           [2007,2008,2009,2010,2011,2012,2013,2014,2015,2016,2017,2018,2019,2020,2021],fontsize=12)

ax1.set_yticks([2,4,6,8,10,12,14,16],fontsize=12,colors='C0')
#ax2.set_yticks([0.5,1.0,1.5,2.0,2.5,3.0,3.5,4.0,4.5],fontsize=12,colors='fuchsia')

for tl in ax1.get_yticklabels():
    tl.set_color(cc)
#for tl in ax2.get_yticklabels():
#    tl.set_color('fuchsia')

ax1.tick_params(axis='y',colors=cc)
#ax2.tick_params(axis='y',colors='fuchsia')  
    
#lns = l1+l2+l3
lns = l1
labs = [l.get_label() for l in lns]
ax1.legend(lns, labs, prop={"size":11},loc = 'best')
plt.savefig('result_3_PARgt0_GPPec_NEEec_ETec_Howland_2007_2020.png')







```
